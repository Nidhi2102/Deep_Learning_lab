```python
import numpy as np

h = "100,150,200,250,300"
p = "20,25,30,35,40"

# Parse the input strings 'h' and 'p'
h = np.array([int(val.strip()) for val in h.split(',')])
p = np.array([int(val.strip()) for val in p.split(',')])

print("h values:", h)
print("p values:", p)
```

```python
# Using np.polyfit for simpler calculation of m and c
# np.polyfit(h, p, 1) returns the coefficients of a polynomial of degree 1 (linear fit)
# where the first coefficient is the slope (m) and the second is the y-intercept (c).

m, c = np.polyfit(h, p, 1)

print(f"Slope (m): {m:.2f}")
print(f"Y-intercept (c): {c:.2f}") # Calculate predictions (p_pred)
p_pred = m * h + c

print("Original p values:", p)
print("Predicted p values:", np.round(p_pred, 2))

from sklearn.metrics import mean_absolute_error, mean_squared_error

# Calculate residual errors
mae = mean_absolute_error(p, p_pred)
mse = mean_squared_error(p, p_pred)
rmse = np.sqrt(mse)

print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```
